```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('diabetes.csv')
```

```python
X=df.drop('Outcome',axis=1)
y=df['Outcome']
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
# Standardize the feature data
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
```

```python
X_train
```

```python
X_test=scaler.transform(X_test)
```

```python
X_test
```

```python
# Initialize Logistic Regression model
model=LogisticRegression()
```

```python
# Fit the model on training data
model.fit(X_train,y_train)
```

```python
# Accuracy on training data
train_accuracy=model.score(X_train, y_train)
print(f"Training Accuracy:{train_accuracy*100:.2f}%")
```

```python
# Accuracy on testing data
test_accuracy=model.score(X_test,y_test)
print(f"Testing Accuracy:{test_accuracy*100:.2f}%")
```

```python
# Function to make predictions
def predict_diabetes(input_data):
#Convert input_data to numpy array and reshape for a single instance
    input_data=np.array(input_data).reshape(1,-1)
    
#Standardize the input data using the same scaler
    input_data=scaler.transform(input_data)
    
#Make prediction
    prediction=model.predict(input_data)
    
    # Output the result
    if prediction[0] == 1:
        return"The person is likely to have diabetes."
    else:
        return"The person is unlikely to have diabetes."
```

```python
# Example input data: [Pregnancies, Glucose, BloodPressure, SkinThickness, Insulin, BMI, DiabetesPedigreeFunction, Age]
new_data = [6,148,72,35,0,33.6,0.627,50]
```

```python
# Call the function
result = predict_diabetes(new_data)
print(result)
```
